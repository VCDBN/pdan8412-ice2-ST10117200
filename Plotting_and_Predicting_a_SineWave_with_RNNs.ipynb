```python
##________________________________Code Attribution______________________________
## The following code was taken from Real Python
## Author: Gruppetta, S.
## Link: https://realpython.com/np-linspace-numpy/

# Using numpy to create a sine wave dataset - 501 data points, between 0 and 50.
x = np.linspace(0, 50, 501)

##__________________end__________________________________________________________
```

```python
##________________________________Code Attribution______________________________
## The following code was taken from NumPy
## Author: NumPy
## Link: https://numpy.org/doc/stable/reference/generated/numpy.sin.html

# Using numpy to calculate the sin function.
y = np.sin(x)

##__________________end__________________________________________________________
```

```python
##________________________________Code Attribution______________________________
## The following code was taken from Matplotlib
## Author: Matplotlib
## Link: https://matplotlib.org/stable/plot_types/basic/plot.html

# Using matplotlib to plot the sine function.
plt.plot(x, y)

# Displays the sine function.
plt.show()

##__________________end__________________________________________________________
```

```python
##________________________________Code Attribution______________________________
## The following code was taken from Scikit Learn
## Author: Scikit Learn
## Link: https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

# Using the train_test_split function to split the data into training and testing data.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.10, shuffle = False)

##__________________end__________________________________________________________
```

```python
##________________________________Code Attribution______________________________
## The following code was taken from Machine Learning Mastery
## Author: Brownlee, J.
## Link: https://machinelearningmastery.com/standardscaler-and-minmaxscaler-transforms-in-python/

# Create an instance of MinMaxScaler.
scaler = MinMaxScaler()

# Using the fit method of the MinMaxScaler on the target variable y_train.
scaler.fit(y_train.reshape(-1, 1))

# Scaling y_train using the MinMaxScaler by reshaping it and then applying the transform method.
y_train_scaled = scaler.transform(y_train.reshape(-1, 1))

# Scaling y_test using the MinMaxScaler by reshaping it and then applying the transform method.
y_test_scaled = scaler.transform(y_test.reshape(-1, 1))

##__________________end__________________________________________________________
```

```python
##________________________________Code Attribution______________________________
## The following code was taken from Machine Learning Mastery
## Author: Brownlee, J.
## Link: https://machinelearningmastery.com/how-to-use-the-timeseriesgenerator-for-time-series-forecasting-in-keras/

# Sets the length of input sequences as 10 for the model.
sequence_length = 10

# Using the TimeseriesGenerator class from the Keras library to generate batches of time series data for training.
train_data_gen = TimeseriesGenerator(y_train_scaled, y_train_scaled, length = sequence_length, batch_size = 1)

##__________________end__________________________________________________________
```

```python
##________________________________Code Attribution______________________________
## The following code was taken from Machine Learning Mastery
## Author: Brownlee, J.
## Link: https://machinelearningmastery.com/understanding-simple-recurrent-neural-networks-in-keras/

# Defining a simple neural network model using the Keras framework.
model = Sequential([SimpleRNN(10, input_shape = (sequence_length, 1), activation = 'relu'), Dense(1)])

# Compiling the model using the Keras framework with the Adam optimizer and mean squared error (MSE) loss function.
model.compile(optimizer = 'adam', loss = 'mse')

##__________________end__________________________________________________________
```

```python
##________________________________Code Attribution______________________________
## The following code was taken from Project Pro
## Author: Vermani, G.
## Link: https://www.projectpro.io/recipes/run-and-fit-data-with-keras-model

# Training the compiled Keras model using the fit method with 5 epochs.
model.fit(train_data_gen, epochs = 5)

##__________________end__________________________________________________________
```

```python
##________________________________Code Attribution______________________________
## The following code was taken from Machine Learning Mastery
## Author: Brownlee, J.
## Link: https://machinelearningmastery.com/display-deep-learning-model-training-history-in-keras/

# Plot the training loss history of the Keras model.
plt.plot(model.history.history['loss'])

# Sets the label for the x-axis of the plot to Epochs.
plt.xlabel("Epochs")

# Sets the label for the y-axis of the plot to Loss.
plt.ylabel("Loss")

# Displays the plot.
plt.show()

##__________________end__________________________________________________________
```

```python
##________________________________Code Attribution______________________________
## The following code was taken from Machine Learning Mastery
## Author: Brownlee, J.
## Link: https://machinelearningmastery.com/how-to-use-the-timeseriesgenerator-for-time-series-forecasting-in-keras/

# Calculating the number of predictions using the trained model on the test data.
num_predictions = len(y_test_scaled) - sequence_length + 1

# Creating a test data generator using the TimeseriesGenerator class from Keras.
# This generator will be used to generate input-output pairs for making predictions on the test data using the trained model.
test_generator = TimeseriesGenerator(y_test_scaled, y_test_scaled, length = sequence_length, batch_size = 1)

# Using the predict method of the trained model to generate predictions for the test data using the test_generator.
predictions = model.predict(test_generator)

# Inverse transform predictions to original scale.
predictions_inv = scaler.inverse_transform(predictions)

# Plotting the actual values of the test target variable against the predictions made by the model using matplotlib.
plt.plot(y_test[sequence_length-1:], label = 'Actual')

# Build the plot by adding the predicted values to the same plot using the variable predictions_inv.
plt.plot(predictions_inv, label = 'Predicted')

# Add a legend to the plot.
plt.legend()

# Displays the plot.
plt.show()

##__________________end__________________________________________________________
```
